## Problem 2. In a room, at least two people in a room of 20 have the same birthday:

In this condition, the reverse is that no two people in the room have the same birthday:

- The 1st person can have the birthday anyday of the year ( p = 1)
- The 2nd person can have the birthday the rest of the year ( p = (365 -1)/365)
- The 3rd person can have the birthday the rest of the year ( p = (365 - 2)/365)

- ...
- The 20th person can have the birthday the rest of the year ( p = (365 - 19)/365)

No people have the same birthday: p(no) = (365-0) /365 *  (365-1) /365 * ... (365-19) /365 = 
At least two people have the same birthday: **1- p(no) = 0.41143838358058027**

```
p = 1
for i in range(20):
  p *= (365 - i)/365  
1- p 
```

## Problem 3. Visualization of the probability of at least two people have the same probability with different number of people in a room

```
import matplotlib.pyplot as plt
res = []

p = 1
for i in range(2, 100):
  p *= (365 - i)/365  
  res.append(1-p)
plt.figure(figsize = (10, 6))
plt.plot(range(2, 100), res)
#plt.plot([res.index(max(res)) + 2], [max(res)], color = "red")
plt.title("The probability of at least two people have the same birthday", size = 18)
plt.xlabel("The number of people in the room", size = 14)
plt.ylabel("The probability ($p$)", size = 14)

plt.axhline(y = 0.5, linestyle = "--", color = "red", xmin=0, xmax=0.24)
plt.axvline(x = 22.2, linestyle = "--", color = "red", ymin=0, ymax=0.5)
plt.text(28.5, 0.48,  '$(23, 0.5)$', horizontalalignment='center')

plt.axhline(y = 0.9, linestyle = "--", color = "blue", xmin=0, xmax=0.41)
plt.axvline(x = 40, linestyle = "--", color = "blue", ymin=0, ymax=0.88)
plt.text(47, 0.88,  '$(40, 0.9)$', horizontalalignment='center')

plt.grid()
plt.show()
  
```
