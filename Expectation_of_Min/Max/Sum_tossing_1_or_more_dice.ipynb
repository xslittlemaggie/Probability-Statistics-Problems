## Part 4: Expectation of the max value when tossing an unbiased dice 1, 2, 3, or more times

### 1. define the function to get the max values

```
def max_simulation(n, k): # k indicates the number of dice
  res = []
  for i in range(n):
    values = [unbiased_dice() for i in range(k)]
    max_val = max(values)
    res.append(max_val)
  return res
```

### 2. visualize the simulaiton of expectation of the min/max

```
def simulation_plot(n, k):
  print("The Max Expectation is: {}".format(np.mean(max_simulation(n, k))))
  plt.hist(max_simulation(n, k))
  plt.title("The histogram of max expectation when tossing {} dice for {} times".format(k, n))
  plt.xlabel("The Expectation")
  plt.ylabel("The Frequency")
  plt.show()
simulation_plot(n=10000, k=2)
```

```
def simulation_mean_plot(n , k = 10): # n indicates number of simulations, k indicates the number of dice, or number of tossing
  mins = []
  maxs = []
  plt.figure(figsize = (10, 6))
  for i in range(1, k):
    mean_min = np.mean(min_simulation(n, k = i))
    mins.append(mean_min)
    mean_max = np.mean(max_simulation(n, k = i))
    maxs.append(mean_max)
  plt.plot(range(1, k), maxs, label = "Max Expectation")
  plt.plot(range(1, k), mins, label = "Min Expectation")
  plt.title("The min/max expectation when simulating {} times with {} dice".format(n, k))
  plt.xlabel("Number of dice")
  plt.ylabel("The Min/Max expectation")
  plt.legend()
  plt.show()
simulation_mean_plot(n= 100000 , k = 10)
```

## Part 5: Expectation of the Sum when tossing an unbiased dice 1, 2, 3, or more times

### 1. define the function to get the sums

```
def sum_simulation(n, k): # k indicates the number of dice
  res = []
  for i in range(n):
    values = [unbiased_dice() for i in range(k)]
    sum_val = sum(values)
    res.append(sum_val)
  return res
```

```
def simulation_plot(n, k):
  print("The Sum Expectation is: {}".format(np.mean(sum_simulation(n, k))))
  plt.hist(sum_simulation(n, k))
  plt.title("The histogram of Sum expectation when tossing {} dice for {} times".format(k, n))
  plt.xlabel("The Expectation")
  plt.ylabel("The Frequency")
  plt.show()
simulation_plot(n=10000, k=2)
```

### 2. visualize the simulaiton of expectation of the min/max *values*

```
def simulation_mean_plot(n , k = 10): # n indicates number of simulations, k indicates the number of dice, or number of tossing
  mins = []
  maxs = []
  sums = []
  plt.figure(figsize = (10, 6))
  for i in range(1, k):
    mean_min = np.mean(min_simulation(n, k = i))
    mins.append(mean_min)
    mean_max = np.mean(max_simulation(n, k = i))
    maxs.append(mean_max)
    mean_sum = np.mean(sum_simulation(n, k = i))
    sums.append(mean_sum)
  plt.plot(range(1, k), maxs, label = "Max Expectation")
  plt.plot(range(1, k), mins, label = "Min Expectation")
  plt.plot(range(1, k), sums, label = "Sum Expectation")
  plt.title("The min/max/sum expectation when simulating {} times with {} dice".format(n, k))
  plt.xlabel("Number of dice")
  plt.ylabel("The Min/Max/Sum expectations")
  plt.legend()
  plt.show()
simulation_mean_plot(n= 100000 , k = 10)
```
